```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import darts as darts
```

```python
#load csv file
df = pd.read_csv('test_nn_c.csv', sep=';')

#show first 5 rows
df.head()
```

```python
# Round every value in the column 2 up to column 19
# Loop through the columns from F1 to F18
for i in range(1, 19):
    column_name = f'F{i}'
    # Convert non-numeric values to NaN
    df[column_name] = pd.to_numeric(df[column_name], errors='coerce')
    # Round the values in the column
    df[column_name] = df[column_name].round(0)

#show first 5 rows
df.head()
```

```python
df.to_csv('test_nn_c_r.csv', sep=';', index=False)
```
